# Replicating the [EDA example](https://github.com/UkeshThapa/Summer-ML-Training/blob/master/exploratory_data_analysis.ipynb?authuser=1) on a differrent data set

We first importing the necessary classes, pandas for data analysis, pyplot for mathematical plots, seahorn for more graphical plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Univarient Analysis

## Importing the dataset : source : [COD_weapons_dataset](https://www.kaggle.com/datasets/shagalsajid/call-of-duty-mobile-weapons-stats)

```python
df = pd.read_csv("/content/drive/MyDrive/AI_folder/Datasets/COD_weapon_data/data.csv")
df.head(5)
```

## Count Plots (Bar Chart)

```python
plt.title("Category Bar Chart")
plt.xticks(rotation=90, ha='right')  # Rotate and align to the right
sns.countplot(data=df, x='category')
df['category'].value_counts()
```

```python
plt.title("Type Bar Chart")
sns.countplot(data=df, x='type')
df['type'].value_counts()
```

## Pie Chart

```python
plt.title("Type Pie Chart")
df['type'].value_counts().plot(kind='pie')
```

## Histogram

```python
plt.title("Accuracy Histogram")
plt.hist(df['Accuracy'])
plt.xlabel("Accuracy rating")
plt.ylabel("Number of weapons")
```

```python
plt.title("Damage Histogram")
plt.hist(df['Damage'])
plt.xlabel("Damage rating")
plt.ylabel("Number of weapons")
```

```python
sns.distplot(df['Accuracy'])
```

```python
sns.distplot(df['Damage'])
```

```python
sns.boxplot(df['Accuracy'])
```

```python
sns.boxplot(df['Damage'])
```

```python
df.describe()
```

# Bivarient Analysis

## Importing the dataset : [Height, Weight and BMI data of 500 People Based on Gender](https://www.kaggle.com/datasets/yersever/500-person-gender-height-weight-bodymassindex)

```python
df2 = pd.read_csv("/content/drive/MyDrive/AI_folder/Datasets/BMI/500_Person_Gender_Height_Weight_Index.csv")
df2.head(5)
```

```python
df2.describe()
```

## Checking for Gender Bias (Disproportionate data causes biasness)

```python
df2['Gender'].value_counts().plot(kind='bar')
```

## Plots

```python
sns.scatterplot(x='Height', y='Weight', hue='Gender', size='Index', data=df2)
plt.show()
```

```python
sns.barplot(x='Index',y='Weight',hue='Gender',data=df2)
```

```python
sns.barplot(x='Index',y='Height',hue='Gender',data=df2)
```

```python
sns.pairplot(df2, hue='Gender')
```

```python
sns.lineplot(x='Index',y='Weight',data=df2)
```
